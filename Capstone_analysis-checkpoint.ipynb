```python
   # Update paths as needed
files = {
    "BNBL": "Historical_Data_BNBL.csv",
    "BBPL": "Historical_Data_BBPL.csv",
    "DPNB": "Historical_Data_DPNB.csv",
    "BIL": "Historical_Data_BIL.csv",
    "BPCL": "Historical_Data_BPCL.csv",
    "DFAL": "Historical_Data_DFAL.csv",
    "DPL": "Historical_Data_DPL.csv",
    "DWAL": "Historical_Data_DWAL.csv",
    "GICB": "Historical_Data_GICB.csv",
    "KCL": "Historical_Data_KCL.csv",
    "PCAL": "Historical_Data_PCAL.csv",
    "RICB": "Historical_Data_RICB.csv",
    "STCB": "Historical_Data_STCB.csv",
    "SVL": "Historical_Data_SVL.csv",
    "TBL": "Historical_Data_TBL.csv",
    "BTCL": "Historical_Data_BTCL.csv"
    
}

# Read the CSVs
dataframes = {}
for name, path in files.items():
    df = pd.read_csv(path)
    df['Company'] = name
    dataframes[name] = df
```

```python
def clean_dataframe(df):
    df.columns = df.iloc[0]  # First row as headers
    df = df[1:]              # Remove original header row
    df = df.rename(columns=lambda x: x.strip())
    df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
    df['Market Price'] = pd.to_numeric(df['Market Price'], errors='coerce')
    df['Company'] = df['Symbol'].iloc[0]
    return df[['Date', 'Market Price', 'Company']].dropna()

# Clean and combine all
cleaned_data = [clean_dataframe(df) for df in dataframes.values()]

merged_df = pd.concat(cleaned_data).sort_values(by=['Company', 'Date']).reset_index(drop=True)
```

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load dividend data
dividend_df = pd.read_csv("Dividend.csv")
dividend_df["Dividend(%)"] = dividend_df["Dividend(%)"].replace('-', '0')
dividend_df["Dividend(%)"] = dividend_df["Dividend(%)"].str.replace('%', '').astype(float)
dividend_df.rename(columns={"Dividend(%)": "Annual_Dividend(%)"}, inplace=True)

# Merge dividend data into merged_df 
merged_df = pd.merge(merged_df, dividend_df, on="Company", how="left")

# Calculate Previous Price for each company 
merged_df["Previous Price"] = merged_df.groupby("Company")["Market Price"].shift(1)

# Estimate per-period Dividend from Annual Dividend (Assuming evenly spread across periods) 
merged_df["Trading_Days"] = merged_df.groupby("Company")["Date"].transform("count")
merged_df["Dividend"] = (merged_df["Annual_Dividend(%)"] / 100 )* merged_df["Previous Price"]
merged_df["Dividend"] = merged_df["Dividend"] / merged_df["Trading_Days"]
# Calculate Total Return (%) using formula 
merged_df["Total Return (%)"] = ((merged_df["Dividend"] + (merged_df["Market Price"] - merged_df["Previous Price"])) / merged_df["Previous Price"]) * 100

# Calculate performance summary 
performance_summary = merged_df.groupby("Company").agg(
    Average_Total_Return=("Total Return (%)", "mean"),
    Volatility=("Total Return (%)", "std"),
    Max_Price=("Market Price", "max"),
    Min_Price=("Market Price", "min"),
    Cumulative_Return=("Total Return (%)", "sum")
).reset_index()

# Display sample output
print("Sample of merged_df with Total Return (%):")
display(merged_df.head())
display(merged_df.tail())

print("Performance Summary:")
display(performance_summary)
```

```python
# Sort by Average Total Return
sorted_perf = performance_summary.sort_values(by="Average_Total_Return", ascending=False).reset_index(drop=True)

# Get top, mid, and low companies
top_company = sorted_perf.iloc[0]
mid_company = sorted_perf.iloc[len(sorted_perf) // 2]
low_company = sorted_perf.iloc[-1]

# Create a summary DataFrame (fix variable names here)
selected_companies = pd.DataFrame([top_company, mid_company, low_company])
selected_companies.index = ['Top Performer', 'Mid Performer', 'Low Performer']

# Display the selected companies
print(selected_companies.round(3))
```

```python

# Extract the company names from the selected rows
top_name = top_company['Company']
mid_name = mid_company['Company']
low_name = low_company['Company']

# Plotting
plt.figure(figsize=(14, 6))

for company_name in [top_name, mid_name, low_name]:
    company_data = merged_df[merged_df["Company"] == company_name]
    plt.plot(company_data["Date"], company_data["Market Price"], label=company_name)

plt.title('Top, Mid, and Low Performing Companies - Market Price Trend')
plt.xlabel('Date')
plt.ylabel('Market Price')
plt.legend(title='Company')
plt.tight_layout()
plt.show()
```

```python
def forecast_arima(company_name, periods=30):
    print(f"Forecasting for: {company_name}")
    df = merged_df[merged_df["Company"] == company_name].copy()
    df = df.set_index("Date").asfreq("D")  # Daily frequency
    df['Market Price'] = df['Market Price'].interpolate()  # Fill missing values

    # Fit ARIMA model
    model = auto_arima(df['Market Price'], seasonal=False, trace=False)
    model_summary = model.summary()
    
    # Forecast
    forecast = model.predict(n_periods=periods)
    forecast_index = pd.date_range(df.index[-1] + pd.Timedelta(days=1), periods=periods, freq='D')
    forecast_series = pd.Series(forecast, index=forecast_index)

    # Plotting
    plt.figure(figsize=(12, 5))
    plt.plot(df['Market Price'], label='Historical')
    plt.plot(forecast_series, label='Forecast', linestyle='--', color='orange')
    plt.title(f"ARIMA Forecast for {company_name}")
    plt.xlabel('Date')
    plt.ylabel('Market Price')
    plt.legend()
    plt.tight_layout()
    plt.show()

    return model_summary
```

```python
from pmdarima import auto_arima
from prophet import Prophet
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
import matplotlib.pyplot as plt
import numpy as np

# Evaluation function
def evaluate_model(true, predicted, model_name, company):
    rmse = np.sqrt(mean_squared_error(true, predicted))
    mae = mean_absolute_error(true, predicted)
    mape = mean_absolute_percentage_error(true, predicted) * 100
    print(f"----- {model_name} Evaluation for {company} -----")
    print(f"RMSE: {rmse:.2f}")
    print(f"MAE: {mae:.2f}")
    print(f"MAPE: {mape:.2f}%\n")

# Prophet Forecast

def prophet_forecast_and_evaluate(df, company_name, periods=30):
    print(f"Running Prophet for {company_name}...")

    df_prophet = df[['Date', 'Market Price']].rename(columns={'Date': 'ds', 'Market Price': 'y'}).dropna()
    model = Prophet(daily_seasonality=True)
    model.fit(df_prophet)
    future = model.make_future_dataframe(periods=periods)
    forecast = model.predict(future)

    forecast_true = df_prophet['y'].values
    forecast_pred = forecast['yhat'][:len(df_prophet)].values
    evaluate_model(forecast_true, forecast_pred, model_name="Prophet", company=company_name)

    # Plot
    fig = model.plot(forecast)
    plt.title(f'Prophet Forecast - {company_name}')
    plt.xlabel("Date")
    plt.ylabel("Market Price")
    plt.show()

# ARIMA Forecast

def arima_forecast_and_evaluate(df, company_name):
    print(f"Running ARIMA for {company_name}...")

    df_arima = df.set_index('Date')['Market Price'].dropna()
    auto_model = auto_arima(df_arima, seasonal=False, stepwise=True, suppress_warnings=True)
    order = auto_model.order
    print(f"Best ARIMA order for {company_name}: {order}")

    model = ARIMA(df_arima, order=order)
    model_fit = model.fit()
    forecast = model_fit.predict(start=0, end=len(df_arima)-1, typ='levels')

    forecast_true = df_arima.values
    forecast_pred = forecast.values
    evaluate_model(forecast_true, forecast_pred, model_name="ARIMA", company=company_name)

    # Plot
    plt.figure(figsize=(10, 5))
    plt.plot(df_arima.index, df_arima.values, label='True')
    plt.plot(df_arima.index, forecast_pred, label='Forecast', linestyle='--')
    plt.title(f'ARIMA Forecast - {company_name}')
    plt.legend()
    plt.show()

# Loop through multiple companies
companies = ['BTCL', 'BNBL', 'PCAL']
for company in companies:
    company_df = merged_df[merged_df['Company'] == company].copy()
    prophet_forecast_and_evaluate(company_df, company)
    arima_forecast_and_evaluate(company_df, company)
```

```python
from prophet import Prophet
from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Evaluation Function
def evaluate_model(true, predicted):
    rmse = np.sqrt(mean_squared_error(true, predicted))
    mae = mean_absolute_error(true, predicted)
    mape = mean_absolute_percentage_error(true, predicted) * 100
    return rmse, mae, mape

# Prophet Forecast
def prophet_forecast(df, periods=30):
    df_prophet = df[['Date', 'Market Price']].rename(columns={'Date': 'ds', 'Market Price': 'y'}).dropna()
    model = Prophet(daily_seasonality=True)
    model.fit(df_prophet)
    future = model.make_future_dataframe(periods=periods)
    forecast = model.predict(future)
    forecast_true = df_prophet['y'].values
    forecast_pred = forecast['yhat'][:len(df_prophet)].values
    return evaluate_model(forecast_true, forecast_pred)

# ARIMA Forecast with Auto Order
def arima_forecast(df):
    df_arima = df.set_index('Date')['Market Price'].dropna()
    auto_model = auto_arima(df_arima, seasonal=False, stepwise=True, suppress_warnings=True)
    model = ARIMA(df_arima, order=auto_model.order)
    model_fit = model.fit()
    forecast = model_fit.predict(start=0, end=len(df_arima)-1, typ='levels')
    return evaluate_model(df_arima.values, forecast.values)

# Compare Prophet and ARIMA for multiple companies
companies = ['BTCL', 'BNBL', 'PCAL']
results = []

for company in companies:
    print(f"Evaluating: {company}")
    df_company = merged_df[merged_df['Company'] == company].copy()
    
    # Prophet
    prophet_rmse, prophet_mae, prophet_mape = prophet_forecast(df_company)
    
    # ARIMA
    arima_rmse, arima_mae, arima_mape = arima_forecast(df_company)
    
    results.append({
        'Company': company,
        'Prophet_RMSE': round(prophet_rmse, 2),
        'ARIMA_RMSE': round(arima_rmse, 2),
        'Prophet_MAE': round(prophet_mae, 2),
        'ARIMA_MAE': round(arima_mae, 2),
        'Prophet_MAPE': round(prophet_mape, 2),
        'ARIMA_MAPE': round(arima_mape, 2),
        'Better_Model': 'Prophet' if prophet_rmse < arima_rmse else 'ARIMA'
    })

# Display comparison
comparison_df = pd.DataFrame(results)
print("\nModel Performance Comparison:")
print(comparison_df)
```

```python
from ipywidgets import interact, Dropdown, DatePicker
import pandas as pd
from prophet import Prophet
import plotly.graph_objects as go
from datetime import timedelta


df['Date'] = pd.to_datetime(df['Date'])

# Company dropdown
company_dropdown = Dropdown(
    options=sorted(df['Company'].unique()),
    description='Company:'
)

# Global min/max dates
min_date = df['Date'].min()
max_date = df['Date'].max()

# Date pickers
start_date_picker = DatePicker(
    description='Start Date:',
    value=min_date,
    disabled=False
)

forecast_date_picker = DatePicker(
    description='Forecast Until:',
    value=min_date + timedelta(days=180),
    disabled=False
)

# Forecasting function
def forecast_with_prophet(df_company, forecast_until):
    df_model = df_company[['Date', 'Market Price']].dropna().rename(columns={'Date': 'ds', 'Market Price': 'y'})
    model = Prophet(daily_seasonality=True)
    model.fit(df_model)

    # Compute number of days to forecast
    last_train_date = df_model['ds'].max()
    forecast_days = (forecast_until - last_train_date).days
    if forecast_days < 1:
        forecast_days = 1

    future = model.make_future_dataframe(periods=forecast_days)
    forecast = model.predict(future)

    fig = go.Figure()
    fig.add_trace(go.Scatter(x=forecast['ds'], y=forecast['yhat'], name='Forecast'))
    fig.add_trace(go.Scatter(x=df_model['ds'], y=df_model['y'], name='Historical'))
    fig.update_layout(title=f"Forecast for {df_company['Company'].iloc[0]} until {forecast_until}", height=450)
    return fig
    
@interact(
    company=company_dropdown,
    start_date=start_date_picker,
    forecast_until=forecast_date_picker
)
def show_forecast(company, start_date, forecast_until):
    if start_date and forecast_until:
        # Convert to pandas Timestamps for consistency
        start_date = pd.Timestamp(start_date)
        forecast_until = pd.Timestamp(forecast_until)

        max_forecast_limit = start_date + pd.Timedelta(days=10000)
        if forecast_until > max_forecast_limit:
            print(" Forecast date exceeds the 10,000-day limit.")
            return

        df_company = df[df['Company'] == company]
        df_filtered = df_company[df_company['Date'] >= start_date]
        if df_filtered.empty:
            print("No data available from the selected start date.")
        else:
            fig = forecast_with_prophet(df_filtered, forecast_until)
            fig.show()

 
        

    

```

```python
# Cell 1: Imports and Data Preparation
import pandas as pd
import plotly.graph_objects as go
from prophet import Prophet
import ipywidgets as widgets
from ipywidgets import interact
import matplotlib.pyplot as plt

# Example subset from your merged_df
df = merged_df[merged_df['Company'].isin(['BTCL', 'PCAL', 'BNBL'])].copy()
df['Date'] = pd.to_datetime(df['Date'])

# Cell 2: Widgets
company_widget = widgets.Dropdown(
    options=['BTCL', 'PCAL', 'DWAL'],
    value='BTCL',
    description='Company:'
)

date_range_widget = widgets.SelectionRangeSlider(
    options=sorted(df['Date'].dt.date.unique()),
    index=(0, len(df['Date'].dt.date.unique()) - 1),
    description='Date Range:',
    layout={'width': '90%'}
)

# Cell 3: Market Price and Volume Plot
@interact(company=company_widget, date_range=date_range_widget)
def plot_price_volume(company, date_range):
    start, end = pd.to_datetime(date_range[0]), pd.to_datetime(date_range[1])
    dff = df[(df['Company'] == company) & (df['Date'].between(start, end))]

    fig = go.Figure()
    fig.add_trace(go.Scatter(x=dff['Date'], y=dff['Market Price'],
                             mode='lines+markers', name='Market Price'))
    fig.add_trace(go.Bar(x=dff['Date'], y=dff['Traded Volume'],
                         name='Traded Volume', yaxis='y2', opacity=0.4))

    fig.update_layout(
        title=f'{company} - Market Price & Volume',
        yaxis=dict(title='Market Price'),
        yaxis2=dict(overlaying='y', side='right', title='Traded Volume'),
        xaxis_title='Date',
        height=500
    )
    fig.show()

# Cell 4: Forecasting with Prophet
@interact(company=company_widget)
def forecast_plot(company):
    dfc = df[df['Company'] == company][['Date', 'Market Price']].rename(columns={'Date': 'ds', 'Market Price': 'y'}).dropna()
    m = Prophet(daily_seasonality=True)
    m.fit(dfc)
    future = m.make_future_dataframe(periods=30)
    forecast = m.predict(future)

    fig = go.Figure()
    fig.add_trace(go.Scatter(x=dfc['ds'], y=dfc['y'], mode='lines', name='Historical'))
    fig.add_trace(go.Scatter(x=forecast['ds'], y=forecast['yhat'], mode='lines', name='Forecast'))
    fig.update_layout(title=f'30-Day Forecast for {company}', xaxis_title='Date', yaxis_title='Market Price')
    fig.show()

```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from pmdarima import auto_arima
import matplotlib.pyplot as plt

def analyze_company(company_name, forecast_periods=30):
    print(f"\n Analyzing Company: {company_name}")
    
    # Prepare the data
    df = merged_df[merged_df["Company"] == company_name].copy()
    df = df.set_index("Date").asfreq("D")
    df["Market Price"] = df["Market Price"].interpolate()  # Fill missing
    
    # Decompose time series
    decomposition = seasonal_decompose(df["Market Price"], model='additive', period=30)
    fig = decomposition.plot()
    fig.set_size_inches(12, 8)
    plt.suptitle(f'Time Series Decomposition - {company_name}', fontsize=16)
    plt.tight_layout()
    plt.show()

    # Check if there's visible seasonality
    seasonal_strength = decomposition.seasonal.std() / df["Market Price"].std()
    print(f" Seasonality strength: {seasonal_strength:.3f}")
    
    if seasonal_strength > 0.1:
        print(f" Applying SARIMA model to {company_name} (seasonality detected)")
        
        model = auto_arima(df["Market Price"], seasonal=True, m=30, trace=False, suppress_warnings=True)
        forecast = model.predict(n_periods=forecast_periods)
        forecast_index = pd.date_range(df.index[-1] + pd.Timedelta(days=1), periods=forecast_periods, freq='D')
        forecast_series = pd.Series(forecast, index=forecast_index)

        # Plot historical + forecast
        plt.figure(figsize=(12, 5))
        plt.plot(df["Market Price"], label='Historical')
        plt.plot(forecast_series, label='SARIMA Forecast', linestyle='--', color='orange')
        plt.title(f"SARIMA Forecast - {company_name}")
        plt.xlabel('Date')
        plt.ylabel('Market Price')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()
    else:
        print(f" No strong seasonality detected in {company_name}. SARIMA not applied.")
```

```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
from datetime import timedelta
```

```python
# Step 2: Load the CSV files
btcl = pd.read_csv('Historical_Data_BTCL.csv')
pcal = pd.read_csv('Historical_Data_PCAL.csv')
bnbl = pd.read_csv('Historical_Data_BNBL.csv')
```

```python
# Preview the data
btcl.head()
```

```python
# Convert 'Date' column to datetime and sort
for df in [btcl, pcal, bnbl]:
    df['Date'] = pd.to_datetime(df['Date'])
    df.sort_values('Date', inplace=True)

# Use Market price for prediction, you can change to 'Open' or 'Volume' etc.
target_col = 'Market Price'
btcl_data = btcl[[target_col]].copy()

# Normalize the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(btcl_data)

# Create sequences
def create_sequences(data, seq_len):
    X, y = [], []
    for i in range(seq_len, len(data)):
        X.append(data[i-seq_len:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

sequence_length = 60
X, y = create_sequences(scaled_data, sequence_length)

# Reshape input to be 3D [samples, time steps, features]
X = np.reshape(X, (X.shape[0], X.shape[1], 1))
```

```python
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X.shape[1], 1)))
model.add(LSTM(units=50))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X, y, epochs=20, batch_size=32)
```

```python
forecast_input = scaled_data[-sequence_length:].reshape(1, sequence_length, 1)
forecast = []

for _ in range(30):
    pred = model.predict(forecast_input)[0][0]
    forecast.append(pred)
    new_input = np.append(forecast_input[0, 1:], [[pred]], axis=0)
    forecast_input = new_input.reshape(1, sequence_length, 1)

# Inverse transform predictions
forecast = scaler.inverse_transform(np.array(forecast).reshape(-1, 1))

# Generate dates for the forecast
last_date = btcl['Date'].max()
forecast_dates = [last_date + timedelta(days=i) for i in range(1, 31)]

# Plot results
plt.figure(figsize=(10, 5))
plt.plot(btcl['Date'].values, btcl[target_col].values.ravel(), label='Historical')
plt.plot(forecast_dates, forecast, label='Forecast (30 days)', linestyle='--')
plt.legend()
plt.xlabel('Date')
plt.ylabel(target_col)
plt.title('LSTM Forecast for BTCL')
plt.show()
```

```python
import pandas as pd
import numpy as np
from prophet import Prophet
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt
import math

# Define evaluation function
def evaluate_model(true, predicted):
    rmse = np.sqrt(mean_squared_error(true, predicted))
    mae = mean_absolute_error(true, predicted)
    mape = mean_absolute_percentage_error(true, predicted) * 100
    return rmse, mae, mape

# Prophet evaluation
def prophet_forecast(df, periods=30):
    df_prophet = df[['Date', 'Market Price']].rename(columns={'Date': 'ds', 'Market Price': 'y'}).dropna()
    model = Prophet(daily_seasonality=True)
    model.fit(df_prophet)
    future = model.make_future_dataframe(periods=periods)
    forecast = model.predict(future)
    true = df_prophet['y'].values
    pred = forecast['yhat'][:len(df_prophet)].values
    return evaluate_model(true, pred)

# ARIMA evaluation
def arima_forecast(df):
    df_arima = df.set_index('Date')['Market Price'].dropna()
    model = ARIMA(df_arima, order=(1,1,1))
    model_fit = model.fit()
    forecast = model_fit.predict(start=0, end=len(df_arima)-1, typ='levels')
    return evaluate_model(df_arima.values, forecast.values)

# LSTM evaluation
def lstm_forecast(df):
    close_prices = df['Market Price'].values
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(close_prices.reshape(-1, 1))

    training_data_len = math.ceil(len(scaled_data) * 0.8)
    train_data = scaled_data[0:training_data_len, :]

    x_train, y_train = [], []
    for i in range(60, len(train_data)):
        x_train.append(train_data[i-60:i, 0])
        y_train.append(train_data[i, 0])
    x_train, y_train = np.array(x_train), np.array(y_train)
    x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))

    model = Sequential()
    model.add(LSTM(100, return_sequences=True, input_shape=(x_train.shape[1], 1)))
    model.add(LSTM(100, return_sequences=False))
    model.add(Dense(25))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(x_train, y_train, batch_size=1, epochs=1, verbose=0)

    test_data = scaled_data[training_data_len - 60:, :]
    x_test = []
    for i in range(60, len(test_data)):
        x_test.append(test_data[i-60:i, 0])
    x_test = np.array(x_test)
    x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

    predictions = model.predict(x_test)
    predictions = scaler.inverse_transform(predictions)

    true = close_prices[training_data_len:]
    pred = predictions.flatten()
    return evaluate_model(true, pred)

# Main evaluation loop
companies = ['BTCL', 'BNBL', 'PCAL']
results = []

for company in companies:
    print(f"Evaluating for: {company}")
    df = merged_df[merged_df['Company'] == company].copy()

    prophet_metrics = prophet_forecast(df)
    arima_metrics = arima_forecast(df)
    lstm_metrics = lstm_forecast(df)

    models = {
        'Prophet': np.mean(prophet_metrics),
        'ARIMA': np.mean(arima_metrics),
        'LSTM': np.mean(lstm_metrics)
    }
    best_model = min(models, key=models.get)

    results.append({
        'Company': company,
        'Prophet_RMSE': prophet_metrics[0],
        'ARIMA_RMSE': arima_metrics[0],
        'LSTM_RMSE': lstm_metrics[0],
        'Prophet_MAE': prophet_metrics[1],
        'ARIMA_MAE': arima_metrics[1],
        'LSTM_MAE': lstm_metrics[1],
        'Prophet_MAPE': prophet_metrics[2],
        'ARIMA_MAPE': arima_metrics[2],
        'LSTM_MAPE': lstm_metrics[2],
        'Best_Model': best_model
    })

# Convert to DataFrame
result_df = pd.DataFrame(results)
print(result_df.round(2))
```
